```python
import shapely
from shapely.geometry import Polygon
import geopandas as gpd
```

```python
def convert_long(longitude):
    if longitude > 180:
        return longitude - 360
    if longitude <= -180:
        return longitude + 360
    return longitude
```

```python
def test_convert_long():
    in_out_expected = {
        190: -170,
        -190: 170,
        360: 0,
        -180: 180,
        180: 180,
        0: 0,
    }

    in_out_actual = {}

    for lon in in_out_expected:
        actual_long = convert_long(lon)
        in_out_actual[lon] = actual_long

    print(in_out_expected)
    print(in_out_actual)
    assert in_out_actual == in_out_expected
```

```python
def polygon_area_m2(polygon, crs):
    geodataframe = gpd.GeoDataFrame(geometry=[polygon], crs=crs)
    geodataframe_projected = geodataframe.to_crs(epsg=3857) # EPSG 3857 is web Mercator CRS which is in units of meters.
    return geodataframe_projected.geometry.area.iloc[0]
```

```python
def test_polygon_area_m2():
    polygon_test = Polygon([
        (0,0),
        (0,100),
        (100,100),
        (100,0)
    ]) # this is for a hypothetical square that is 100m x 100m
    area = polygon_area_m2(polygon_test, "EPSG:3857")
    print(area)
    assert area == 10_000
```

```python
test_polygon_area_m2()
```
